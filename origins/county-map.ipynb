```python
import geopandas as gpd
import pandas as pd
import json
```

```python
counties = gpd.read_file("../unsynced-data/county-crops-v1.gpkg")
counties = counties.to_crs("EPSG:4326")
```

```python
cdl_codes = pd.read_csv("../unsynced-data/cdl-codes.csv")
```

```python
# for every column that begins with crop_ change the values to the value divided by the max value in that column
for col in counties.columns:
    if col.startswith("crop_"):
        counties[col] = counties[col] / counties[col].max()
```

```python
# return max value of every column that begins with crop_
max_crop = []
for col in counties.columns:
    if col.startswith("crop_"):
        max_crop.append(counties[col].max())
print (max_crop)
```

```python
# write to geojson File
counties.to_file("../unsynced-data/county-crops-percentage.geojson", driver="GeoJSON")
```
