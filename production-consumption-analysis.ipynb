```python
import pandas as pd
import numpy as np
```

```python
# read in production and consumption data
food = pd.read_csv('./output-data/county-population-consumption-production-scaled.csv')
# create a blank pd dataframe called food_summary
food_list = []
# for every column in food if the column name includes "total_kcal_consumed_" add that column name to food_list
for column in food:
    if "total_kcal_consumed_" in column:
        food_list.append(column)
# for every value in food_list remove total_kcal_consumed_ string
for i in range(len(food_list)):
    food_list[i] = food_list[i].replace("total_kcal_consumed_", "")
# create a dataframe called food_summary that has one column called crop that is the contents of food_list and one column that is max_consumed_kcal that is just the value 0
food_summary = pd.DataFrame({'crop': food_list, 'max_produced_county':np.nan, 'max_produced_kcal': 0, 'min_produced_county':np.nan, 'min_produced_kcal':0, 'max_consumed_county':np.nan, 'max_consumed_kcal': 0, 'min_consumed_county':np.nan, 'min_consumed_kcal':0})
# for each row in food_summary, find the max value in the column crop in the food dataframe and add that value to the max_produced_county column in food_summary
for index, row in food_summary.iterrows():
    food_summary.loc[index, 'max_produced_county'] = food.loc[food['kcal_produced_' + row['crop']] == food['kcal_produced_' + row['crop']].max(), 'geographic_area_name'].values[0]
    food_summary.loc[index, 'max_produced_kcal'] = food['kcal_produced_' + row['crop']].max()
    food_summary.loc[index, 'min_produced_county'] = food.loc[food['kcal_produced_' + row['crop']] == food['kcal_produced_' + row['crop']].min(), 'geographic_area_name'].values[0]
    food_summary.loc[index, 'min_produced_kcal'] = food['kcal_produced_' + row['crop']].min()
    food_summary.loc[index, 'max_consumed_county'] = food.loc[food['total_kcal_consumed_' + row['crop']] == food['total_kcal_consumed_' + row['crop']].max(), 'geographic_area_name'].values[0]
    food_summary.loc[index, 'max_consumed_kcal'] = food['total_kcal_consumed_' + row['crop']].max()
    food_summary.loc[index, 'min_consumed_county'] = food.loc[food['total_kcal_consumed_' + row['crop']] == food['total_kcal_consumed_' + row['crop']].min(), 'geographic_area_name'].values[0]
    food_summary.loc[index, 'min_consumed_kcal'] = food['total_kcal_consumed_' + row['crop']].min()
# remove the word dried from the crop column
food_summary['crop'] = food_summary['crop'].str.replace('dried', '')
food_summary['crop'] = food_summary['crop'].str.replace(' flour', '')
food_summary
```

```python
# write food_summary to a csv
food_summary.to_csv('./output-data/production-consumption-max-min.csv', index=False)
```

```python
# make a dataframe of the food column geographic_area_name and all columns in food that begin with kcal_produced
food_produced = food[['geographic_area_name'] + [col for col in food.columns if col.startswith('kcal_produced')]]
# return the row with the most non 0 values
food_produced['max_produced_county'] = food_produced.apply(lambda x: x[x != 0].count(), axis=1)
```

```python
# return the 5 rows with the max value in the max_produced_county column
food_produced.nlargest(5, 'max_produced_county')
```
